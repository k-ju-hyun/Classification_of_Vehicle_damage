```python
from tensorflow.keras.preprocessing.image import load_img
from tensorflow.keras.preprocessing.image import img_to_array

import numpy as np
import matplotlib.pyplot as plt
```

```python
train_path = '/content/drive/MyDrive/Datasets/copy_images/trainset'
images = []

for i in img_train_list:
    img_tr_path = os.path.join(train_path, i)
    img = load_img(img_tr_path, target_size = (280, 280))
    img = img_to_array(img)

    images.append(img)

img_tr_arr = np.array(images)
img_tr_arr.shape
```

#### Y : One-Hot Encoding

```python
import numpy as np
from tensorflow.keras.utils import to_categorical

class_n = len(np.unique(y_train))

y_train = to_categorical(y_train, class_n)
y_test = to_categorical(y_test, class_n)
y_valid = to_categorical(y_valid, class_n)
```

#### 1) 구조 설계

```python
import tensorflow as tf
from tensorflow import keras

from tensorflow.keras.backend import clear_session
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input, Dense, Flatten, Conv2D, MaxPool2D, BatchNormalization, Dropout

from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import EarlyStopping
```

```python
## Sequential API

# 세션 클리어
clear_session()

# 모델 선언
model = keras.models.Sequential()

# 모델 블록 조립
model.add( Input(shape = (280, 280, 3)) )

model.add( Conv2D(filters = 32,
                                padding = 'same',
                                kernel_size = (3, 3),
                                strides = (1, 1),
                                activation = 'relu') )
model.add( BatchNormalization() )

model.add( Conv2D(filters = 32,
                                padding = 'same',
                                kernel_size = (3, 3),
                                strides = (1, 1),
                                activation = 'relu') )
model.add( BatchNormalization() )

model.add( MaxPool2D( pool_size = (2, 2),
                                        strides = (2, 2)) )
# model.add( Dropout(0.2) )

model.add( Conv2D(filters =  64,
                                kernel_size = (3, 3),
                                strides = (1, 1),
                                padding = 'same',
                                activation = 'relu') )
model.add( BatchNormalization() )

model.add( Conv2D(filters = 64,
                                kernel_size = (3, 3),
                                strides = (1, 1),
                                padding = 'same',
                                activation = 'relu') )
model.add( BatchNormalization() )

model.add( MaxPool2D( pool_size = (2, 2),
                                        strides = (2, 2)) )
# model.add( Dropout(0.25) )

model.add( Conv2D(filters =  128,
                                kernel_size = (3, 3),
                                strides = (1, 1),
                                padding = 'same',
                                activation = 'relu') )
model.add( Conv2D(filters = 128,
                                kernel_size = (3, 3),
                                strides = (1, 1),
                                padding = 'same',
                                activation = 'relu') )
model.add( BatchNormalization() )
# model.add( Dropout(0.3) )

model.add( MaxPool2D( pool_size = (2, 2),
                                        strides = (2, 2)) )

model.add( Conv2D(filters =  256,
                                kernel_size = (3, 3),
                                strides = (1, 1),
                                padding = 'same',
                                activation = 'relu') )
model.add( Conv2D(filters = 256,
                                kernel_size = (3, 3),
                                strides = (1, 1),
                                padding = 'same',
                                activation = 'relu') )
model.add( BatchNormalization() )

model.add( MaxPool2D( pool_size = (2, 2),
                                        strides = (2, 2)) )
model.add( Dropout(0.3) )

model.add( Flatten() )
# model.add( Dense(512, activation = 'relu') )
model.add( Dense(256, activation = 'relu') )
model.add( Dense(128, activation = 'relu') )
model.add( BatchNormalization() )

model.add( Dense(2, activation = 'sigmoid') )

# 컴파일
model.compile(loss = 'binary_crossentropy', metrics = ['accuracy'],
                    optimizer = 'adam')

# 요약
model.summary()
```

#### 2) 학습
* EarlyStopping 설정하고 학습시키기

```python
from tensorflow.keras.callbacks import EarlyStopping
```

```python
es = EarlyStopping(monitor = 'val_loss',
                             min_delta = 0,
                             patience = 10,
                             verbose = 1,
                             restore_best_weights = True)
```

```python
history = model.fit(train_x, y_train, epochs = 10000, batch_size = 32,
                            validation_data = (valid_x, y_valid), verbose = 1, callbacks = [es])
```

#### 3) test set으로 예측하고 평가하기

```python
y_pred = model.predict(img_te_arr)
y_pred.shape
```

```python
print(y_test.shape)
single_test_y = y_test.argmax(axis = 1)
print(single_test_y.shape)
```

```python
print(confusion_matrix(single_test_y, single_y_pred))
print(classification_report(single_test_y, single_y_pred))
```

```python
## Functional API
# 세션 클리어
clear_session()

# 레이어 사슬처럼 엮기
il = Input( shape = (280, 280, 3) )

hl = Conv2D(filters = 32,
                 kernel_size = (3, 3),
                 strides = (1, 1),
                 padding = 'same',
                 activation = 'relu')(il)
hl = Conv2D(filters = 32,
                 kernel_size = (3, 3),
                 strides = (1, 1),
                 padding = 'same',
                 activation = 'relu')(hl)
hl = BatchNormalization()(hl)
hl = MaxPool2D(pool_size = (2, 2),
                         strides = (2, 2))(hl)
hl = Dropout(0.25)(hl)

hl = Conv2D(filters = 64,
                 kernel_size = (3, 3),
                 strides = (1, 1),
                 padding = 'same',
                 activation = 'relu')(hl)
hl = Conv2D(filters = 64,
                 kernel_size = (3, 3),
                 strides = (1, 1),
                 padding = 'same',
                 activation = 'relu')(hl)
hl = BatchNormalization()(hl)
hl = MaxPool2D(pool_size = (2, 2),
                         strides = (2, 2))(hl)
hl = Dropout(0.25)(hl)

hl = Conv2D(filters = 128,
                 kernel_size = (3, 3),
                 strides = (1, 1),
                 padding = 'same',
                 activation = 'relu')(hl)
hl = Conv2D(filters = 128,
                 kernel_size = (3, 3),
                 strides = (1, 1),
                 padding = 'same',
                 activation = 'relu')(hl)
hl = BatchNormalization()(hl)
hl = MaxPool2D(pool_size = (2, 2),
                         strides = (2, 2))(hl)
hl = Dropout(0.25)(hl)

hl = Flatten()(hl)
hl = Dense(1024, activation = 'relu')(hl)
# hl = Dense(512, activation = 'relu')(hl)
hl = BatchNormalization()(hl)
hl = Dropout(0.35)(hl)
ol = Dense(2, activation = 'sigmoid')(hl)

# 모델의 시작과 끝 지정
model = Model(il, ol)

# 컴파일
model.compile(loss = 'binary_crossentropy', metrics = ['accuracy'],
                        optimizer = Adam())

# 요약
model.summary()
```

```python
import tensorflow as tf
from tensorflow import keras

from tensorflow.keras.backend import clear_session
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input, Dense, Flatten, Conv2D, MaxPool2D, BatchNormalization, Dropout

from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import EarlyStopping
```

```python
## VGG 16

# 세션 클리어
clear_session()

# 레이어 사슬처럼 엮기
il = Input( shape = (280, 280, 3) )

hl = Conv2D(filters = 64,
                    kernel_size = (3, 3),
                    padding = 'same',
                    activation = 'relu')(il)
hl = Conv2D(filters = 64,
                    kernel_size = (3, 3),
                    padding = 'same',
                    activation = 'relu')(hl)
hl = MaxPool2D(pool_size = (2, 2))(hl)
# hl = BatchNormalization()(hl)

hl = Conv2D(filters = 128,
                    kernel_size = (3, 3),
                    padding = 'same',
                    activation = 'relu')(hl)
hl = Conv2D(filters = 128,
                    kernel_size = (3, 3),
                    padding = 'same',
                    activation = 'relu')(hl)
hl = MaxPool2D(pool_size = (2, 2))(hl)
# hl = BatchNormalization()(hl)

hl = Conv2D(filters = 256,
                    kernel_size = (3, 3),
                    padding = 'same',
                    activation = 'relu')(hl)
hl = Conv2D(filters = 256,
                    kernel_size = (3, 3),
                    padding = 'same',
                    activation = 'relu')(hl)
hl = MaxPool2D(pool_size = (2, 2))(hl)
# hl = BatchNormalization()(hl)

hl = Conv2D(filters = 512,
                    kernel_size = (3, 3),
                    padding = 'same',
                    activation = 'relu')(hl)
hl = Conv2D(filters = 512,
                    kernel_size = (3, 3),
                    padding = 'same',
                    activation = 'relu')(hl)
hl = MaxPool2D(pool_size = (2, 2))(hl)
# hl = BatchNormalization()(hl)

hl = Conv2D(filters = 512,
                    kernel_size = (3, 3),
                    padding = 'same',
                    activation = 'relu')(hl)
hl = Conv2D(filters = 512,
                    kernel_size = (3, 3),
                    padding = 'same',
                    activation = 'relu')(hl)
hl = MaxPool2D(pool_size = (2, 2))(hl)
hl = BatchNormalization()(hl)

hl = Flatten()(hl)
hl = Dense(4096, activation = 'relu')(hl)
hl = Dense(4096, activation = 'relu')(hl)
hl = Dropout(0.35)(hl)

ol = Dense(2, activation = 'sigmoid')(hl)

# 모델의 시작과 끝 지정
model = Model(il, ol)

# 컴파일
model.compile(loss = 'binary_crossentropy', metrics = ['accuracy'],
                        optimizer = Adam())

# 요약
model.summary()
```

```python
from tensorflow.keras.applications import VGG16
import tensorflow as tf
from tensorflow import keras

from tensorflow.keras import layers
from tensorflow.keras.backend import clear_session
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input, Dense, Flatten, Conv2D, MaxPool2D, BatchNormalization, Dropout

from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import EarlyStopping

from keras.models import load_model
```

```python
train_x = img_tr_arr
test_x = img_te_arr
valid_x = img_valid_arr
```

```python
mean_n = img_tr_arr.mean()
std_n = img_tr_arr.std()

train_x = (img_tr_arr - mean_n) / std_n
test_x = (img_te_arr - mean_n) / std_n
valid_x = (img_valid_arr - mean_n) / std_n
```

#### 1) VGG16 불러와서 저장하기
* include_top=False로 설정하여 분류기를 제외하고 미리 학습된 가중치 imagenet을 로드
* .trainable을 True로 설정하여 모델의 모든 레이어들이 fine-tuning에 대해 업데이트


#### 2) VGG16과 연결한 구조 설계

```python
clear_session()

model_path = '/content/drive/MyDrive/first_model.h5'

base_model = load_model(model_path)

vgg16 = VGG16(input_shape = (280, 280, 3),
                        include_top = False,
                        weights = 'imagenet')

for layer in vgg16.layers:
    layer.trainable = True

x = vgg16.output
x = Flatten()(x)
x = Dense(512, activation = 'relu')(x)
# x = Dense(1024, activation = 'relu')(x)
x = Dropout(0.5)(x)
x = Dense(2, activation = 'sigmoid')(x)

model_vgg = Model(vgg16.input, x)

model_vgg.compile(loss = 'binary_crossentropy', metrics = ['accuracy'],
                                optimizer = 'adam')

model_vgg.summary()
```

#### 3) 학습

```python
from tensorflow.keras.callbacks import EarlyStopping

es = EarlyStopping(monitor = 'val_loss',
                             min_delta = 0,
                             patience = 10,
                             verbose = 1,
                             restore_best_weights = True)
```

```python
history = model_vgg.fit(train_x, y_train, epochs = 10000, batch_size = 32,
                                    validation_data = (valid_x, y_valid), verbose = 1, callbacks = [es])
```

#### 4) 성능 평가

```python
from sklearn.metrics import confusion_matrix, classification_report
```

```python
y_pred = model_vgg.predict(test_x)
single_y_pred = y_pred.argmax(axis = 1)
single_test_y = y_test.argmax(axis = 1)
```
